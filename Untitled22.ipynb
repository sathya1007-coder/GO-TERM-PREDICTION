```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

# Load the dataset
file_path = "/content/SST_SELECTED_HOURLY.xlsx"
sheet_name = 'sst12n90e'  # Update the sheet name if needed
dataframe = pd.read_excel(file_path, sheet_name=sheet_name, usecols=[0, 1, 2], dtype=str)

# Convert YYYYMMDD and HHMM columns to datetime
dataframe['YYYYMMDD'] = dataframe['YYYYMMDD'].astype(str)
dataframe['HHMM'] = dataframe['HHMM'].astype(str).str.zfill(4)  # Ensure HHMM is always 4 digits
dataframe['Datetime'] = pd.to_datetime(dataframe['YYYYMMDD'] + dataframe['HHMM'], format='%Y%m%d%H%M')

# Drop old columns
dataframe.drop(columns=['YYYYMMDD', 'HHMM'], inplace=True)

# Convert SST to float
dataframe['SST'] = dataframe['SST'].astype(float)

# Set Datetime as index
dataframe.set_index('Datetime', inplace=True)

# Split into train and test sets
X = dataframe['SST'].values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()

# Walk-forward validation
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print(f'predicted={yhat:.6f}, expected={obs:.6f}')

# Evaluate forecasts
rmse = sqrt(mean_squared_error(test, predictions))
print(f'Test RMSE: {rmse:.3f}')

# Plot forecasts against actual outcomes
plt.figure(figsize=(12, 6))
plt.plot(dataframe.index[size:], test, label='Actual SST')
plt.plot(dataframe.index[size:], predictions, color='red', label='Predicted SST')
plt.xlabel('Datetime')
plt.ylabel('SST')
plt.title('ARIMA Model Forecast vs Actual SST')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load the dataset
file_path = "/content/SST_SELECTED_HOURLY.xlsx"
sheet_name = 'sst12n90e'  # Update the sheet name if needed
dataframe = pd.read_excel(file_path, sheet_name=sheet_name, usecols=[0, 1, 2], dtype=str)

# Convert YYYYMMDD and HHMM columns to datetime
dataframe['YYYYMMDD'] = dataframe['YYYYMMDD'].astype(str)
dataframe['HHMM'] = dataframe['HHMM'].astype(str).str.zfill(4)  # Ensure HHMM is always 4 digits
dataframe['Datetime'] = pd.to_datetime(dataframe['YYYYMMDD'] + dataframe['HHMM'], format='%Y%m%d%H%M')

# Drop old columns
dataframe.drop(columns=['YYYYMMDD', 'HHMM'], inplace=True)

# Convert SST to float
dataframe['SST'] = dataframe['SST'].astype(float)

# Set Datetime as index
dataframe.set_index('Datetime', inplace=True)

# Extract values
dates = dataframe.index
values = dataframe['SST'].values.astype('float32')

# Identify the last 2 unique days
date_unique = pd.Series(dataframe.index.date).unique()
test_dates = date_unique[-2:]

# Create masks for training and testing
test_mask = pd.Series(dataframe.index.date).isin(test_dates).values
train_mask = ~test_mask

# Split data into train and test
train, test = values[train_mask], values[test_mask]
dates_train, dates_test = dates[train_mask], dates[test_mask]

# Normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train.reshape(-1, 1))  # Fit on train
test_scaled = scaler.transform(test.reshape(-1, 1))        # Transform test

# Walk-forward validation (Fixed)
history = [x for x in train_scaled]
predictions = []
for t in range(len(test_scaled)):
    model = ARIMA(history, order=(5,1,0), enforce_stationarity=False, enforce_invertibility=False)
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]  # Forecasted value
    predictions.append(yhat)

    history.append([yhat])  # Append predicted value instead of actual test value
    print(f'predicted={yhat:.6f}, expected={test_scaled[t][0]:.6f}')

# Inverse transform predictions and test set to original scale
test_inversed = scaler.inverse_transform(test_scaled)
predictions_inversed = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

# Evaluate forecasts
rmse = sqrt(mean_squared_error(test_inversed, predictions_inversed))
print(f'Test RMSE: {rmse:.3f}')

# Plot forecasts against actual outcomes
plt.figure(figsize=(12, 6))
plt.plot(dates_test, test_inversed, label='Actual SST')
plt.plot(dates_test, predictions_inversed, color='red', label='Predicted SST')
plt.xlabel('Datetime')
plt.ylabel('SST')
plt.title('ARIMA Model Forecast vs Actual SST')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
